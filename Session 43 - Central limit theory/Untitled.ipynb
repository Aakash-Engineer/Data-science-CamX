```python
import scipy
import numpy as np
import pandas as pd
import seaborn as sns
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
normal = stats.lognorm.rvs(0.9, loc=25, scale=5, size=1000)
normal
sns.kdeplot(normal)
```

```python
fig, ax = plt.subplots(ncols=2, nrows=1, figsize=(16, 4))

sns.kdeplot(normal, ax=ax[0])
sns.kdeplot(np.log(normal), ax=ax[1])
```

```python
sample_size = 1000
sampling_dist = []

for i in range(50):
    sample = np.random.choice(normal, sample_size, replace=False)
    sampling_dist.append(sample.mean())

print('Sampling mean   :', np.mean(sampling_dist))
print('Population mean :', normal.mean())
```
